```python
%load_ext autoreload
%autoreload 2
import numpy as np
import matplotlib.pyplot as plt
import nsta.core as nc
import nsta.helpers as nh 
plt.rcParams.update({'font.size': 15})
```

```python
x,y,z = np.random.rand(3)
num_steps = 1000
dt = 0.01
lorenz_jacs = []
xyz = np.zeros((num_steps,3))

for t in range(num_steps):

    #get lorenz jacobian at that time-step, store
    jac = nh.lorenz_jac(x,y,z,dt)
    lorenz_jacs.append(jac)

    #do lorenz update
    x_up,y_up,z_up = nh.lorenz_update(x,y,z)

    xyz[t] = [x,y,z]
    
    x+= dt*x_up
    y+= dt*y_up
    z+= dt*z_up

lorenz_jacs = np.stack(lorenz_jacs)
lorenz_jacs = np.expand_dims(lorenz_jacs,axis = 0)


    

```

```python
lle_estimate_lorenz = nc.estimate_LLE(lorenz_jacs, p = 100)
```
